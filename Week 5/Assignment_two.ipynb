# The statistical and probability tests in the diferent scenario's 

Shapiro-Wilk'stest

This is used to check whether a dataset follows normal distribution.
- This can be used to check if the population  density across different regions follows normal distribution which enables an analyist to detremine which statistical methods to use for further analysis.
- IT can also be used to assess whether the distribution of household sizes in different regions is normally distributed.

One-way ANOVA test

This is used to determine the means of three or more independent groups to determine if they are statistically significant amongst themselves.
- One-way ANOVA testis used to compare the average household size across different regions.

Pearson’s correlation test

This measures the strength and direction of a linear relationship between two continous variables.
- In the case of the Ugandan Population Census, Pearson's correlation test is used to analyze tthe realtionship between population density and average household size across the different regions thus helping us to understand whether densely populated regions have siginficantly different household sizes compared to less populated ones.

The Cumulative distribution function
- Cumulative distribution function is used to determine the proportion of the population that falls below a certain age.
- It can also be used to determine the probability of a household having the most number of members.

Poisson distribution

Poisson distribution is used to model count-based data and events occurring in a fixed interval.
- Poisson distribution can be used to modle the number of births per month in a given district assuming a relatively constant birth rate.
- It can also be used to model the number of people moving into or out of the region within a fixed period.

# The probability of a student being "Enrolled" under the variable "Target"? 

```python
#Filtering the datset to include only rows where the "Target" column has the value 'Enrolled'
Enrolled = data[data['Target'] == 'Enrolled']['Target'].count()
#View results
print(Enrolled)
```

```python
# Count the total number in the 'Target' column
Enrolled_total = data['Target'].count()
# View Results
print(f'The total number of students that enrolled is:{Enrolled_total}')
```

```python
# Calculation of the probability of a student being enrolled
Probability_of_student_being_enrolled = Enrolled/Enrolled_total
#view Results
print(f'The probability of a student being enrolled is:{Probability_of_student_being_enrolled}')
```
